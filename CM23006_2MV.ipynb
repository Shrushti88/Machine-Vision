```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()

filename = next(iter(uploaded))
img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
scaled = cv2.resize(img, None, fx=1.5, fy=1.5)

plt.imshow(scaled)
plt.title("Scaled Image")
plt.axis("off")
plt.show()
```

```python
(h, w) = img.shape[:2]
center = (w//2, h//2)

M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))

plt.imshow(rotated)
plt.title("Rotated Image")
plt.axis("off")
plt.show()
```

```python
rows, cols = img.shape[:2]

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[20,80],[220,40],[80,220]])

matrix = cv2.getAffineTransform(pts1, pts2)
affine = cv2.warpAffine(img, matrix, (cols, rows))

plt.imshow(affine)
plt.title("Affine Transformation")
plt.axis("off")
plt.show()
```

```python
import numpy as np

# Example coordinates (x, y)
p1 = np.array([100, 120])
p2 = np.array([350, 120])

pixel_distance = np.linalg.norm(p1 - p2)

print("Pixel Distance:", round(pixel_distance, 2), "pixels")
```

```python
pixel_to_mm = 0.05

distance_mm = pixel_distance * pixel_to_mm

print("Estimated Distance =", round(distance_mm,2), "mm")
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(img)
plt.title("Original")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(scaled)
plt.title("Scaled")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(rotated)
plt.title("Rotated")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(affine)
plt.title("Affine")
plt.axis("off")

plt.tight_layout()
plt.show()
```
